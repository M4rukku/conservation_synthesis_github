```python
issn_prefix = {}
for ind, row in data.iterrows():
    issn = row['issn']
    
    if row["doi"] is None:
        continue
    prefix = doi_prefix_pattern.search(row["doi"]).group(1)
    if issn is not None:
        stored = issn_prefix.get(issn, None)
        if stored is not None:
            if stored==prefix:
                continue
            else:
                print(f"Conflixt: {stored} != {prefix}")
        else:
            issn_prefix[issn] = prefix
```
